```python
import dotenv
dotenv.load_dotenv()

from db import load_test
import os
import google.generativeai as genai
import prompt_templates
import wiki
import rich

GOOGLE_API_KEY = os.environ.get("GEMINI_API_KEY")
genai.configure(api_key=GOOGLE_API_KEY)

wiki = wiki.Wiki("LotmWiki")
```

```python
docstore = load_test(chunk_size=2000)
```

```python
prev_chunks = [node.text for node in docstore.nodes[3:4]]
curr_chunk = docstore.nodes[4].text

prev_chunks = prompt_templates.apply_prev_chunks_template(prev_chunks)

user_msg = prompt_templates.DEFAULT_PROMPT_TEMPLATE.format(
    prev_chunks=prev_chunks,
    curr_chunk=curr_chunk,
)
rich.print(user_msg)
```

```python
model = genai.GenerativeModel(
        "models/gemini-1.5-flash",
        tools=wiki.functions.values(),
        system_instruction=prompt_templates.DEFAULT_SYSTEM_MSG,
    )
```

```python
response = model.generate_content(user_msg)
rich.print(response)
```

```python
from enum import Enum
import dataclasses
import wiki
from section_examples.characters import Personality

@dataclasses.dataclass
class CharacterAttributes(wiki.EntityAttributes):

    # Class-level attributes_list
    attributes_list: list[wiki.Attribute] = [
        Personality(
            name=Personality.name(),
            type=str,
            default="",
            description=Personality.description(),
            update_every_n_insertions=5,
        ),
        # TODO: add other attributes.
    ]

    # Class-level attributes dictionary
    attributes = {attr.name: attr for attr in attributes_list}

    # Class-level AttributeNames Enum
    AttributeNames = Enum(
        "AttributeNames", {attr.name: attr.name for attr in attributes_list}
    )

    def __post_init__(self):
        # Instance-level attributes dictionary (if needed)
        self.attributes = self.__class__.attributes

    def get_attributes(self) -> dict[str, wiki.Attribute]:
        return self.attributes
```
